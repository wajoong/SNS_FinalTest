```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
import sys
import numpy  
import pandas as pd
import os
import xlwt

keyword  = input('크롤링할 키워드를 입력해주세요.:')
start_date =  input('조회 시작일자를 입력하세요.(예:2020/01/01):')
end_date = input('조회 종료일자를 입력하세요.(예:2020/05/01, 시작일로부터 6개월 이내로만 입력):')
f_dir = input("3.결과 파일을 저장할 폴더명을 작성하세요.(예:c:\\data\\):")
n = time.localtime()
s = '%04d-%02d-%02d-%02d-%02d-%02d' % (n.tm_year, n.tm_mon, n.tm_mday, n.tm_hour, n.tm_min, n.tm_sec)
os.makedirs(f_dir+s+'-'+keyword)
os.chdir(f_dir+s+'-'+keyword)
f_txt=f_dir+s+'-'+keyword+'\\'+s+'-'+keyword+'.txt'
f_csv=f_dir+s+'-'+keyword+'\\'+s+'-'+keyword+'.csv'
f_xls=f_dir+s+'-'+keyword+'\\'+s+'-'+keyword+'.xls'

path = "c:/temp/chromedriver_win32/chromedriver.exe"
driver = webdriver.Chrome(path)
driver.get('https://www.g2b.go.kr/index.jsp')
time.sleep(1)

search = driver.find_element(By.ID,"bidNm")
search.send_keys(keyword)
start_d= driver.find_element(By.ID,"fromBidDt")
start_d.clear()
start_d.send_keys(start_date)
end_d = driver.find_element(By.ID,"toBidDt")
end_d.clear()
end_d.send_keys(end_date)
driver.find_element(By.CLASS_NAME,'btn_dark').click()
time.sleep(1)

driver.switch_to.frame('sub')
driver.switch_to.frame('main')
table = driver.find_element(By.CLASS_NAME, 'table_list_tbidTbl')
tbody = table.find_element(By.TAG_NAME, 'tbody')
rows = tbody.find_elements(By.TAG_NAME,'tr')

n = 1
nl = []
c1 = []
c2 = []
c3 = []
c4 = []
c5 = []
c6 = []
c7 = []
c8 = []
c9 = []
c10 = []
c11 = []

for index, row in enumerate(rows):
    content = row.find_elements(By.TAG_NAME, 'td')
    nl.append(n)
    c1.append(content[0].text)
    c2.append(content[1].text)
    c3.append(content[2].text)
    temp = content[3].find_element(By.TAG_NAME, 'div').find_element(By.TAG_NAME, 'a')
    c4.append(temp.text)
    c5.append(temp.get_attribute("href"))
    c6.append(content[4].text)
    c7.append(content[5].text)
    c8.append(content[6].text)
    c9.append(content[7].text)
    c10.append(content[8].text)
    c11.append(content[9].text)
    n += 1

resultFrame = pd.DataFrame()
resultFrame['용역']=c1
resultFrame['공고번호-차수']=c2
resultFrame['분류']=c3
resultFrame['공고명']=c4
resultFrame['URL']=c5
resultFrame['공고기관']=c6
resultFrame['수요기관']=c7
resultFrame['계약방법']=c8
resultFrame['입력일시']=c9
resultFrame['공동수급']=c10
resultFrame['통찰여부']=c11

f = open(f_txt, 'a',encoding='UTF-8')
for i in range(len(nl)):
    f.write(str(nl[i]) + '번째 공고내용입니다. ~~~~\n')
    f.write('1.업무:' + c1[i] + '\n')
    f.write('2.공고번호-차수:' + c2[i] + '\n')
    f.write('3.분류:' + c3[i] + '\n')
    f.write('4.공고명:' + c4[i] + '\n')
    f.write('5.URL:' + c5[i] + '\n')
    f.write('6.공고기관:' + c6[i] + '\n')
    f.write('7.수요기관:' + c7[i] + '\n')
    f.write('8.계약방법:' + c8[i] + '\n')
    f.write('9.입력일시(입찰마감일시):' + c9[i] + '\n')
    f.write('10.공동수급:' + c10[i] + '\n')
    f.write('11.투찰여부:' + c11[i] + '\n\n\n')
f.close()
resultFrame.to_csv(f_csv, encoding="utf-8-sig") 
resultFrame.to_excel(f_xls)
print("요청하신 데이터 추출이 성공적으로 끝났습니다.")
```
